```python
import pickle
import itertools
import numpy as np
from scipy import spatial
```

```python
# Load the overall embeddings received from Pyke

entire_embeddings = pickle.load( open( "Pyke50_pickle.p", "rb" ) )
```

```python


# Creating a dictionary with only the item embeddings


# dictionary of item embeddings
item_embedddings = {} 
for key, val in entire_embeddings.items(): 
    if key.startswith('http://hm2.com/article#'):
        
        
        item_embedddings.update( {key : val} )
          
print(len(item_embedddings))
pickle.dump( item_embedddings, open( "item_embedddings.p", "wb" ) )
```

```python
## need to delete later

item_embedddings = pickle.load( open( "item_embedddings.p", "rb" ) )
```

```python
## take a slice of item embeddings for testing

sample_item_embedddings = dict(itertools.islice(item_embedddings.items(), 15))
#sample_item_embedddings.keys()
print(sample_item_embedddings)
```

```python
## Load output embeddings
output_embeddings = pickle.load( open( "sample_output_embeddings.p", "rb" ) )
sample_output_embeddings = list(itertools.chain.from_iterable(output_embeddings))
print(sample_output_embeddings)
```

```python
## this cell calculates the cosine similarity among y and y_hats
final_cosine = []


for yhat in sample_output_embeddings:
    local_cosine = []
    for k,v in sample_item_embedddings.items():

        result = 1 - spatial.distance.cosine(yhat, v)
        local_cosine.append(result)
    final_cosine.append(local_cosine)

final_cosine


```

```python
## this cell ranks the urls in decending order and lists the top n most frequent elemnts
ranks = []
for i in final_cosine:
    numpy_cosine = np.array(i)
    rank_list = numpy_cosine.argsort()[::-1][:10]
    ranks.append(rank_list)

ranks
```

```python
# this cell captures y_hat predictions
y_hat = []
for lists in ranks:
    local_key = []
    for key in lists:
        first_key = list(sample_item_embedddings)[key]
        local_key.append(first_key)
    y_hat.append(local_key)
#y_hat
```

```python
## load original the keys of y
y = []
original_sample_output_embeddings = pickle.load( open( "original_sample_output_embeddings.p", "rb" ) ).tolist()
#print(len(original_sample_output_embeddings))
for i in original_sample_output_embeddings:
    for k,v in item_embedddings.items():
        if i ==v:
            y.append(k)

#y
    
    
```

```python
y.pop(1)
y.append('http://hm2.com/article#685816016199003')
print(y)
```

```python
## This cell is to generate the matrix for final learning
y_arr = [None]* len(y)
y_hat_arr = [None]* len(y_hat)
y_arr = y
y_hat_arr = y_hat

i = 0
count = 0
while(i<len(y_arr)):
    if (y_arr[i] in y_hat_arr[i]):
        count +=1
    i +=1
Ratio = count/len(y_arr)
print('Learning Ratio is : ',Ratio)
```
